## Remove special characters from gtfs_pt_vehicles.add.xml

This step is mandatory as the special characters created will not permit the launch of the simulation.

```python


with open('gtfs_pt_vehicles.add.xml', 'r') as file:
    data = file.read()

# Replace all occurrences of ="(' with =
data = data.replace('="(\'', '="')

# Replace all occurrences of ,)" with "
data = data.replace(',)"', '"')

# Replace all occurrences of ,).0" with "
data = data.replace(',).0"', '"')


# Remove any residual single quotes at the end of lines
data = data.replace('\'', '')

with open('gtfs_pt_vehicles.add2.xml', 'w') as file:
    file.write(data)


```

## Any attribute

When pedestrians have to take a public transportation, only one id is specified in the trips file: this resulted in more than ten thousand pedestrians being stuck at the stops at the end of the simulation because they had not been able to take that particular transportation in time. By modifying the file with the ANY attribute, the pedestrian will take any other trip of the same line will take him to the next edge described in the trip. By doing this, there are less pedestrians left on the stops at the end of the simulation.

```python
import xml.etree.ElementTree as ET

# Parsing the XML file
tree = ET.parse('pubblici_orari.routes.xml')
root = tree.getroot()

# Iterating over all ride elements
for ride in root.findall('.//ride'):
    # Replace the content of the lines attribute with "ANY"
    ride.set('lines', 'ANY')

# Save the modifications to the file
tree.write('pubblici_orari.routes_modified.xml', xml_declaration=True, encoding='utf-8')


```

```python
with open('gtfs_pt_vehicles.add.xml', 'r') as file:
    data = file.read()

# Sostituisci tutte le occorrenze di ="(' con =
data = data.replace('="(\'', '="')

# Sostituisci tutte le occorrenze di ,)" con "
data = data.replace(',)"', '"')

# Rimuovi eventuali apici singoli residui alla fine delle righe
data = data.replace('\'', '')

with open('gtfs_pt_vehicles.add2.xml', 'w') as file:
    file.write(data)

```

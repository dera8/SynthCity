# Merge dei file gtfs

```python
import pandas as pd
csv_stops = pd.read_csv("gtfs/stops.txt", sep=',', dtype = 'unicode')
csv_stops.head()
```

```python
csv_stop_times = pd.read_csv("gtfs/stop_times.txt", sep=',', dtype = 'unicode')
csv_stop_times.head()
```

```python
csv_trips = pd.read_csv("gtfs/trips.txt", sep=',', dtype = 'unicode')
csv_trips.head()
```

```python
csv_stop_times_trips = csv_stop_times.merge(csv_trips, on=['vehicle_id'])
csv_stop_times_trips.sort_values(["vehicle_id"], ascending = True)
```

```python
csv_stoptimes_stops = csv_stop_times.merge(csv_stops, on=["stop_id"])
csv_stoptimes_stops.sort_values(["vehicle_id"], ascending = True)
```

```python
csv_stoptimes_stops2 = csv_stoptimes_stops[['vehicle_id','arrival_time','departure_time','stop_id','stop_name','stop_sequence','stop_lat','stop_lon']]
csv_stoptimes_stops2.sort_values(["vehicle_id","stop_id","stop_sequence"], ascending = True)
```

```python
csv_stopout = pd.read_csv("stops_1aprile.out.csv", sep=';', dtype = 'unicode')
csv_stopout.rename(columns = {'stopinfo_busStop':'busStop_id'}, inplace = True)
csv_stopout.rename(columns = {'stopinfo_id':'vehicle_id'}, inplace = True)
csv_stopout.sort_values(["vehicle_id"], ascending = True).head()
```

```python
csv_stopout2 = csv_stopout[['busStop_id','stopinfo_delay','stopinfo_ended','vehicle_id','stopinfo_initialPersons','stopinfo_loadedPersons','stopinfo_started','stopinfo_type','stopinfo_unloadedPersons']]
csv_stopout2
```

```python
csv_stopout2.to_csv(r'csv_stopout2.csv', index = False, sep=';')
```

```python
csv_stopout2_no0 = pd.read_csv("csv_stopout2.csv", sep=';', dtype = 'unicode')
```

```python
csv_gtfs_stops = pd.read_csv("gtfs_pt_stops.add_filtered.csv", sep=';', dtype = 'unicode')
csv_gtfs_stops
```

```python
csv_stops_stopout2no0 = csv_gtfs_stops.merge(csv_stopout2_no0, on=["busStop_id"])
csv_stops_stopout2no0.sort_values(["vehicle_id"], ascending = True)
```

```python
csv_stopout3 = csv_stoptimes_stops2.merge(csv_stops_stopout2no0, on=["vehicle_id","stop_name"])
csv_stopout3.sort_values(["vehicle_id","stop_sequence"], ascending = True)
```

```python
routes = pd.read_csv("gtfs/trips.txt", sep=',', dtype = 'unicode')
routes
```

```python
csv_stopout4 = csv_stopout3.merge(routes, on=["vehicle_id"])
csv_stopout4.sort_values(["vehicle_id"], ascending = True)
```

```python
csv_stopout4.to_csv(r'dataset_sintetico_1aprile_coordinates.csv', index = False, sep=';')
```

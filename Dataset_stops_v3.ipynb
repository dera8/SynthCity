## Dopo aver creato il dataset, facciamo un'analisi dello stesso

```python
import numpy as np
import pandas as pd
from datetime import datetime
import statistics 
data=pd.read_csv("dataset_sintetico.csv", sep=';', dtype = 'unicode')
data['arrival_time']=pd.to_datetime(data["arrival_time"],format='%H:%M:%S')#,errors='coerce')
data['stopinfo_started']=pd.to_datetime(data["stopinfo_started"],format='%H:%M:%S')#errors='coerce')



data['sumo_delay'] = (data['stopinfo_started'] - data['arrival_time']).dt.total_seconds()
#data['sumo_delay']
data['sumo_delay_modified'] = np.where((data.sumo_delay > 80000.0), data.sumo_delay-86400.0, data.sumo_delay)  #quando un termine della sottrazione supera la mezzanotte, la differenza porta un giorno in più. Ho risolto così al momento ma la soluzione migliore è aggiungere il giorno all'orario.
data['Delay'] = np.where((data.sumo_delay_modified <-80000.0), data.sumo_delay_modified+86400.0, data.sumo_delay_modified)
data_ordered=data['Delay'].sort_values()

```
